### Célula 1 — imports e diretórios

```python
import sys
from pathlib import Path
import deeplabcut
import tqdm

ROOT_NOTEBOOK = Path.cwd().resolve().parent
if str(ROOT_NOTEBOOK) not in sys.path:
    sys.path.append(str(ROOT_NOTEBOOK))

from src.paths import ROOT_DIR, VIDEOS_DIR, PROJECT_DIR, CONFIG_PATH, VIDEO_SWEEP, ensure_basic_dirs, show_paths

ensure_basic_dirs()
show_paths()

PROJECT = "bab_bar_only"
EXPERIMENTER = "Edo"  # seu nome ou usuário

assert VIDEO_SWEEP.exists(), f"Vídeo não encontrado: {VIDEO_SWEEP}"
```

### Célula 2 — criar projeto novo

```python
# Rode apenas se quiser criar um projeto novo.
# Se o projeto já existe, pule esta célula.

config_path = deeplabcut.create_new_project(
    PROJECT,
    EXPERIMENTER,
    [str(VIDEO_SWEEP)],
    working_directory=str(ROOT_DIR),
    copy_videos=True,
    multianimal=False
)

print(config_path)
```

### Célula 3 — usar config existente

```python
# Use esta célula quando o projeto já existir
assert CONFIG_PATH.exists(), f"config.yaml não encontrado: {CONFIG_PATH}"
print(CONFIG_PATH)
```

### Célula 4 — editar config

```python
from deeplabcut.utils import auxiliaryfunctions

cfg = auxiliaryfunctions.read_config(CONFIG_PATH)

# Modifique aqui apenas se quiser mudar os keypoints
cfg["bodyparts"] = [
    "beam_left",
    "beam_center",
    "beam_right",
]

# Modifique aqui apenas se quiser alterar a quantidade de frames
cfg["numframes2pick"] = 50

auxiliaryfunctions.write_config(CONFIG_PATH, cfg)

print("bodyparts:", cfg["bodyparts"])
print("numframes2pick:", cfg["numframes2pick"])
```
