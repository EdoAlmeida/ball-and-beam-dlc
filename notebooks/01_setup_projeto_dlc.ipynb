# 01 - DeepLabCut Project Setup

This notebook creates the DLC project and adjusts the `config.yaml` file.

**Execute only once**, or when it is necessary to recreate the project.

### Cell 1 — Imports and Directories

```python
import sys
from pathlib import Path
import deeplabcut
import tqdm

ROOT_NOTEBOOK = Path.cwd().resolve().parent
if str(ROOT_NOTEBOOK) not in sys.path:
    sys.path.append(str(ROOT_NOTEBOOK))

from src.paths import ROOT_DIR, VIDEOS_DIR, PROJECT_DIR, CONFIG_PATH, VIDEO_SWEEP, ensure_basic_dirs, show_paths

ensure_basic_dirs()
show_paths()

PROJECT = "bab_bar_only"
EXPERIMENTER = "Edo"  # your name or username

assert VIDEO_SWEEP.exists(), f"Video not found: {VIDEO_SWEEP}"
```

### Cell 2 — Create New Project

```python
# Run only if you want to create a new project.
# If the project already exists, skip this cell.

config_path = deeplabcut.create_new_project(
    PROJECT,
    EXPERIMENTER,
    [str(VIDEO_SWEEP)],
    working_directory=str(ROOT_DIR),
    copy_videos=True,
    multianimal=False
)

print(config_path)
```

### Cell 3 — Use Existing Config

```python
# Use this cell when the project already exists
assert CONFIG_PATH.exists(), f"config.yaml not found: {CONFIG_PATH}"
print(CONFIG_PATH)
```

### Cell 4 — Edit Config

```python
from deeplabcut.utils import auxiliaryfunctions

cfg = auxiliaryfunctions.read_config(CONFIG_PATH)

# Modify here only if you want to change the keypoints
cfg["bodyparts"] = [
    "beam_left",
    "beam_center",
    "beam_right",
]

# Modify here only if you want to change the number of frames
cfg["numframes2pick"] = 50

auxiliaryfunctions.write_config(CONFIG_PATH, cfg)

print("bodyparts:", cfg["bodyparts"])
print("numframes2pick:", cfg["numframes2pick"])
```
