```python
from moviepy import VideoFileClip
import os

video_original = r"C:\Users\edoal\Downloads\rampa_positiva.mov"
video_convertido = r"C:\Users\edoal\Downloads\rampa_positiva.mp4"

print("🔄 Iniciando conversão para H.264 (alta qualidade)...")

try:
    clip = VideoFileClip(video_original)

    clip.write_videofile(
        video_convertido,
        codec="libx264",
        audio_codec="aac",
        preset="slow",          # melhor compressão mantendo qualidade
        bitrate=None,
        ffmpeg_params=[
            "-crf", "16",       # 16 = quase sem perda (18 também é ótimo)
            "-pix_fmt", "yuv420p",
            "-profile:v", "high",
            "-level", "4.0",
            "-movflags", "+faststart"
        ]
    )

    clip.close()
    print(f"✅ Sucesso! Vídeo pronto em: {video_convertido}")

except Exception as e:
    print(f"❌ Erro na conversão: {e}")
```
